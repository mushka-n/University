# Задание 2

```python
from scipy.cluster.hierarchy import *
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
from scipy.spatial.distance import pdist, squareform
```

```python
# Евклидово расстояние
distances = pdist(data_normalized)

# Объединение элементов в кластер
linked = linkage(distances, method='ward')
```

```python
# Метод локтя

end = linked[-10:, 2]
end_ = end[::-1]
idxs = np.arange(1, len(end) + 1)
plt.plot(idxs, end_)

acc = np.diff(end, 2)  
acc_ = acc[::-1]
result = acc_.argmax() + 2 
plt.plot(idxs[:-2] + 1, acc_)
plt.show()

print("По методу локтя рекомендуется раздение на", result, "кластеров" )
```

Также по графику видно что разделения на 6 и на 8 также хорошо подходят

По моему мнению, 2 - слишком маленькое кол-во, так что я возьму среднее значение - 6

```python
clusters=fcluster(linked, 6, criterion='maxclust')
[sum(clusters==x) for x in range(1,7)]
silhouette_score(data_normalized, clusters)
print("Оценка силуэта для кластеризации на 6 -", silhouette_score(data_normalized, clusters))
```

Итог:

```python
ward_method = linkage(data_normalized, method='ward')
plt.figure(figsize=(15, 10))
dendrogram(ward_method, color_threshold=2.3,truncate_mode="lastp",p=20)
plt.axhline(y=2.3, c='k')
plt.show()
labels_hi = fcluster(ward_method, t=6, criterion='maxclust')
```

```python
tsne = TSNE(random_state=17)
tsne_rep = tsne.fit_transform(data_normalized)

plt.figure(figsize =(15, 10))
plt.scatter(tsne_rep[:, 0], 
            tsne_rep[:, 1],
            c = labels_hi)
plt.show()
```

# Задание 3

```python
from sklearn.cluster import DBSCAN
```

```python
def my_score(arr):
    res = 0
    i = 0
    for x in arr:
        res -= ((-1) ** i) * x
        i += 1
    return abs(res)

array = np.arange(0.01,0.5,0.02)
results = [0,0,0,0,[]]
my_results = [0,0,0,9999,[]]
print("Общие результаты: ")
for eps in array:
    dbscan = DBSCAN(eps=eps, min_samples=2)
    labels_db = dbscan.fit_predict(data_normalized)
    cluster_sum = [sum(labels_db==x) for x in np.unique(labels_db)]
    if (3 < len(cluster_sum) < 10 ):
        score = silhouette_score(data_normalized, labels_db)
        myscore = my_score(cluster_sum)
        
        if score > results[2]:
            results = [eps,len(cluster_sum), score, myscore, cluster_sum]
            
        if myscore < my_results[3]:
            my_results = [eps,len(cluster_sum), score, myscore, cluster_sum]
            
        print('Eps - ',round(eps,3), 
              ' |  clusters num - ', len(cluster_sum), 
              ' |  score -',round(score,5),
              ' |  myscore -', myscore)
    
print('\nЛучший результат:')
print('Eps - ',round(results[0],3), '\nclustersnum - ', results[1], '\nscore -',results[2], '\nmy score -',results[3])
print(*results[4])
        
print('\nЛучший по моей схеме сравнения:')
print('Eps - ',round(my_results[0],3), '\nclustersnum - ', my_results[1], '\nscore -',my_results[2], '\nmy score -',my_results[3])
print(*my_results[4])

   
```

Видно, что данные малопригодны для кластеризации по плотности


Из всех eps нам по образу силуэта наиболее подходит 0.37, однако в нем слишком велики различия между кластерами

```python
dbscan = DBSCAN(eps=0.37, min_samples=2)
labels_db = dbscan.fit_predict(data_normalized)

fig = plt.figure(figsize=(15, 10))
fig.subplots_adjust(hspace=.5, wspace=.2)
sns.scatterplot(tsne_rep[:,0],
                tsne_rep[:,1],
                hue=["cluster-{}".format(h) for h in labels_db])
plt.show()
```

При eps равной 0.23 хотя бы 2 кластера становятся примерно равны друг другу, однако для успешной кластеризации этого мало

```python
dbscan = DBSCAN(eps=0.23, min_samples=2)
labels_db = dbscan.fit_predict(data_normalized)

fig = plt.figure(figsize=(15, 10))
fig.subplots_adjust(hspace=.5, wspace=.2)
sns.scatterplot(tsne_rep[:,0],
                tsne_rep[:,1],
                hue=["cluster-{}".format(h) for h in labels_db])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
plt.figure(figsize=(15, 10))
scores = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, 
                    init = 'random',
                    n_init = 500,)
    kmeans.fit(data_normalized)
    labels_km = kmeans.predict(data_normalized)
    scores.append(silhouette_score(data_normalized, labels_km))

plt.plot(range(2, 10), scores, 'x-r')
plt.axhline(y=scores[2], color='g', linestyle='-')
plt.show()
```
